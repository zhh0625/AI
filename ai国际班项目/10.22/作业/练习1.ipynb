
Using simple NumPy operations for manipulating images
=====================================================

This script illustrates how to use basic NumPy operations, such as slicing,
masking and fancy indexing, in order to modify the pixel values of an image.



```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt

camera = data.camera()
#在此处修改camera图像的部分像素值
# print(camera)
# print(camera[100:])
camera[510:] = 255
# print(camera<100)
camera[camera<100] = 255

#此处设置阈值生成掩膜
x = np.arange(len(camera))
# print(10*x%255)
y = (10*x)%255
camera[x,y] = 0



#此处生成黑色圆圈掩膜，并设置掩膜像素值, 然后将其作用到camera图像上

x,y = camera.shape[0],camera.shape[1]
X,Y = np.ogrid[:x,:y]
print(camera.shape)
F = (X-0.5*x)**2+(Y-0.5*y)**2>(0.5*x)**2
C = (X-0.5*x)**2/250**2+(Y-0.5*y)**2/200**2>1
camera[C] = 0


plt.figure(figsize=(4, 4))
plt.imshow(camera, cmap='gray')
plt.axis('off')
plt.show()
```
